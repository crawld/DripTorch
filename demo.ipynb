### Burn unit and wind direction

Load the GeoJSON file as a DripTorch burn unit object and specify a wind direction from the west

```python

burn_unit = dt.BurnUnit.from_json(geojson_1, wind_direction=0)
firing_area = burn_unit.buffer_control_line(5)
firing_area = firing_area.buffer_downwind(20)
blackline_area = burn_unit.difference(firing_area)
```

### Igniters and ignition crew

```python
dash_igniter = dt.Igniter(1,dash_length=.5)
point_crew = dt.IgnitionCrew.clone_igniter(dash_igniter, 7)
```

### Firing techniques

```python

technique = dt.firing.Back(firing_area, point_crew)
pattern = technique.generate_pattern(10,clockwise=False)





map = dt.Map(burn_unit)
map.add_firing_area(firing_area)
map.add_blackline_area(blackline_area)
map.add_pattern(pattern)
map.show()
```

```python
qf_ignition_file = pattern.to_quicfire(burn_unit, filename='./quicfire_output_test.dat', time_offset=0, resolution=1)
print(qf_ignition_file)

#print(pattern.to_json())
```
