```python
import driptorch

geojson = {"type":"FeatureCollection","features":[{"type":"Feature","properties":{},"geometry":{"type":"Polygon","coordinates":[[[-114.44869995117188,47.088504171925706],[-114.44470882415771,47.08745225315146],[-114.44342136383057,47.09066638416644],[-114.44496631622313,47.09236102969754],[-114.44633960723877,47.0924194647886],[-114.45281982421875,47.089205439567344],[-114.45153236389159,47.08815353464254],[-114.44869995117188,47.088504171925706]]]}}]}
  
burn_unit = driptorch.BurnUnit.from_json(geojson, 270)
firing_area = burn_unit.buffer_control_line(10)
firing_area = firing_area.buffer_downwind(25)
blackline_area = burn_unit.difference(firing_area)

slow_igniter = driptorch.Igniter(2, 0)
fast_igniter = driptorch.Igniter(2, 0)

crew = driptorch.IgnitionCrew.clone_igniter(fast_igniter, 4)
crew = driptorch.IgnitionCrew.from_list([fast_igniter, slow_igniter], same_rate = False)

strip = driptorch.FiringTechniques.back(firing_area, crew)
strip_pattern = strip.generate_pattern(20)

ignition_file = strip_pattern.to_quicfire('test.inp')
print(ignition_file)

map = driptorch.Map(burn_unit)
map.add_firing_area(firing_area)
map.add_blackline_area(blackline_area)
map.add_pattern(strip_pattern)
map.show()

```
