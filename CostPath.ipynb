```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from skimage.draw import bezier_curve
from scipy.ndimage import distance_transform_edt
from cost_distance import cost_distance_analysis_numpy as cda
```

```python
# Initailize binary image
logical_map = np.zeros((800, 800), dtype=bool)

# Create a curve across the top of the domain and rasterize
rr, cc = bezier_curve(100, 100, 200, 200, 100, 400, 1, shape=(800, 800))
logical_map[rr, cc] = 1
rr, cc = bezier_curve(100, 400, 0, 600, 100, 700, 1, shape=(800, 800))
logical_map[rr, cc] = 1
plt.imshow(logical_map, cmap=plt.cm.gray)
plt.show()

# Compute euclidean distance transform
distance_map = distance_transform_edt(~logical_map)
distance_map = np.round(distance_map, 0).astype(int)
plt.imshow(distance_map, cmap=plt.cm.gray)
plt.show()

# Identify iso-distance contours in 40 meter intervals
level_sets = np.zeros((800, 800), dtype=bool)
level_sets[distance_map % 40 == 0] = 1
plt.imshow(level_sets, cmap=plt.cm.gray)
plt.show()
```
